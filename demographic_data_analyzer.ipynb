```python
import pandas as pd
import numpy as np

# Mengatur agar hasil random tetap sama setiap kali dijalankan
np.random.seed(42)

def generate_dummy_data(n_rows=1000):
    data = {
        'age': np.random.randint(17, 90, size=n_rows),
        'workclass': np.random.choice(['Private', 'Self-emp-not-inc', 'Self-emp-inc', 'Federal-gov', 'Local-gov', 'State-gov', 'Without-pay', 'Never-worked'], n_rows),
        'fnlwgt': np.random.randint(20000, 1500000, size=n_rows),
        'education': np.random.choice(['Bachelors', 'Some-college', '11th', 'HS-grad', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', '9th', '7th-8th', '12th', 'Masters', '1st-4th', '10th', 'Doctorate', '5th-6th', 'Preschool'], n_rows),
        'education-num': np.random.randint(1, 17, size=n_rows),
        'marital-status': np.random.choice(['Married-civ-spouse', 'Divorced', 'Never-married', 'Separated', 'Widowed', 'Married-spouse-absent', 'Married-AF-spouse'], n_rows),
        'occupation': np.random.choice(['Tech-support', 'Craft-repair', 'Other-service', 'Sales', 'Exec-managerial', 'Prof-specialty', 'Handlers-cleaners', 'Machine-op-inspct', 'Adm-clerical', 'Farming-fishing', 'Transport-moving', 'Priv-house-serv', 'Protective-serv', 'Armed-Forces'], n_rows),
        'relationship': np.random.choice(['Wife', 'Own-child', 'Husband', 'Not-in-family', 'Other-relative', 'Unmarried'], n_rows),
        'race': np.random.choice(['White', 'Asian-Pac-Islander', 'Amer-Indian-Eskimo', 'Other', 'Black'], n_rows),
        'sex': np.random.choice(['Male', 'Female'], n_rows),
        'capital-gain': np.random.randint(0, 100000, size=n_rows),
        'capital-loss': np.random.randint(0, 5000, size=n_rows),
        'hours-per-week': np.random.randint(1, 99, size=n_rows),
        'native-country': np.random.choice(['United-States', 'Cambodia', 'England', 'Puerto-Rico', 'Canada', 'Germany', 'India', 'Japan', 'Mexico', 'Philippines'], n_rows),
        'salary': np.random.choice(['<=50K', '>50K'], n_rows)
    }
    
    return pd.DataFrame(data)

# Membuat dataframe
df_dummy = generate_dummy_data(1000)

# Menampilkan 5 data teratas
print(df_dummy.head())

# Simpan ke CSV jika ingin digunakan di file lain
df_dummy.to_csv('adult.data.csv', index=False)
```

```python
# Contoh tes cepat: Berapa rata-rata usia pria di data dummy ini?
avg_men = df_dummy[df_dummy['sex'] == 'Male']['age'].mean()
print(f"Rata-rata usia pria: {avg_men:.1f}")
```

### calculate_demographic_data

```python
race_count = df_dummy['race'].value_counts()
print(race_count)
```

```python
average_age_men = round(df_dummy[df_dummy['sex'] == 'Male']['age'].mean(), 1)
print(average_age_men)
```

```python
percentage_bachelors = round(len(df_dummy[df_dummy['education'] == 'Bachelors']) / len(df_dummy) * 100, 1)
print(percentage_bachelors)
```

```python
higher_education = df_dummy[df_dummy['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
higher_education_rich = round((higher_education['salary'] == '>50K').mean() * 100, 1)
print(higher_education_rich)
```

```python
min_work_hours = df_dummy['hours-per-week'].min()
print(min_work_hours)
```

```python
num_min_workers = df_dummy[df_dummy['hours-per-week'] == min_work_hours]
rich_percentage = round((num_min_workers['salary'] == '>50K').mean() * 100, 1)
print(rich_percentage)
```

```python
# Hitung total orang per negara
country_count = df_dummy['native-country'].value_counts()
# Hitung orang bergaji >50K per negara
country_rich_count = df_dummy[df_dummy['salary'] == '>50K']['native-country'].value_counts()

# Hitung persentasenya
highest_earning_country_percentage = round((country_rich_count / country_count * 100).max(), 1)
highest_earning_country = (country_rich_count / country_count * 100).idxmax()
```

```python
top_IN_occupation = df_dummy[(df_dummy['native-country'] == 'India') & (df_dummy['salary'] == '>50K')]['occupation'].value_counts().idxmax()
print(top_IN_occupation)
```
